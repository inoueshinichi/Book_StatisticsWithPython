```python
import numpy as np
import pandas as pd

%precision 3
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 折れ線グラフ
x = np.arange(0, 10)
y = np.array([2,3,4,3,5,4,6,7,4,8])
plt.plot(x, y, color='black')
plt.title("lineplot matplotlib")
plt.xlabel('x')
plt.ylabel('y')
# グラフの保存
plt.savefig("./data/simpleline.jpg")
plt.show()
```

```python
# seabornのヒストグラム
fish_data = np.array([2,3,3,4,4,4,4,5,5,6])
fish_data
```

```python
sns.distplot(fish_data, bins=5, color="black", kde=False)
```

```python
# bins = 1
sns.distplot(fish_data, bins=1, color="black", kde=False)
```

```python
# カーネル密度推定ON
# カーネル密度推定はヒストグラムの平滑化の手法の一つ
sns.distplot(fish_data, color="black", kde=True)
```

```python
# 2変量データに対するヒストグラム
fish_multi = pd.read_csv("./data/3-3-2-fish_multi_2.csv")
fish_multi
```

```python
fish_multi.groupby('species').describe()
```

```python
# 種類ごとに対象データを取り出して変数に格納
length_a = fish_multi.query('species == "A"')['length']
length_b = fish_multi.query('species == "B"')['length']
```

```python
sns.distplot(length_a, bins=5, color='black', kde=False)
sns.distplot(length_b, bins=5, color='gray', kde=False)
```

```python
# 多変量データ用箱ひげ図
sns.boxplot(x="species", y="length", data=fish_multi)
```

```python
# violinplot
sns.violinplot(x="species", y="length", data=fish_multi)
```

```python
# 棒グラフ
sns.barplot(x="species", y='length', data=fish_multi, color='gray')
```
