```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats
% precision 3
% matplotlib inline
```

```python
# 真の分布
# モデル   ビールの売上~N(beta0+beta1*気温, 0.5^2)
# beta0 = 35.0
# beta1 = 1.0

beta0 = 35.0
beta1 = 1.0
var = 0.5 ** 2
temperature = 20
mean = beta0 + beta1 * temperature

correct_dist = stats.norm(loc=mean, scale=var)
corr_samples = correct_dist.rvs(size=100)
corr_samples_mean = sp.mean(corr_samples)
corr_samples_var = sp.var(corr_samples)

#print(corr_samples)
print(corr_samples_mean)
print(corr_samples_var)

# 尤度
corr_likelihood = stats.norm.pdf(x=corr_samples, loc=mean, scale=var)
print("真の分布のlog尤度：\n", sp.log(corr_likelihood))

corr_likelihood_new = stats.norm.logpdf(x=corr_samples, loc=mean, scale=var)
print("真の分布の尤度(logpdf):", corr_likelihood_new)
```

```python
# 推定された分布
# beta0 = 34.6102
# beta1 = 0.7654
gbeta0 = 35
gbeta1 = 0.998 # 変更点　この値を変えて、推定分布の形状と真の分布の平均対数尤度を確認する。　２つの分布形状が近いほど、平均対数尤度は高い
gvar = 0.5 ** 2
gtemp = 20
gmean = gbeta0 + gbeta1 * gtemp
g_dist = stats.norm(loc=gmean, scale=gvar)
g_samples = g_dist.rvs(size=100)

#print(g_samples)

#  真の分布から得られた応答変数yに対する推定分布の尤度
g_likelihood = stats.norm.pdf(x=corr_samples, loc=gmean, scale=gvar)
print("真の分布から得られた応答変数yに対する推定分布の尤度：\n", g_likelihood)
g_loglikelihood = sp.log(g_likelihood)
print("上記の対数尤度:\n", g_loglikelihood)

g_likelihood_new = stats.norm.logpdf(x=corr_samples, loc=gmean, scale=gvar)
print("推定分布の対数尤度(logpdf):", g_likelihood_new)
```

```python
# 平均対数尤度
print(sp.sum(corr_likelihood * g_loglikelihood))

#  相対エントロピー
print(sp.sum(corr_likelihood * sp.log(corr_likelihood)) - sp.sum(corr_likelihood * g_loglikelihood))
```

```python
# 真の分布同士の平均対数尤度
print(sp.sum(corr_likelihood * sp.log(corr_likelihood)))

# 真の分布同士の相対エントロピー
print(sp.sum(corr_likelihood * sp.log(corr_likelihood)) - sp.sum(corr_likelihood * sp.log(corr_likelihood)))
```

```python
"""説明変数Xiも含む場合"""
# 観測値は(Xi, Yi)の組になる

# 真の分布
# Xi : 気温、 Yi : ビールの売上
# モデル   ビールの売上~N(beta0+beta1*Xi, 0.5^2)
# beta0 = 34.6102
# beta1 = 0.7654

corr_beta0 = 34.6102
corr_beta1 = 0.7654

# 真の分布からサンプリングシミュレーション

# 気温(0~40℃)
temp = stats.uniform(0.0, 40.0).rvs(size=100)

# ビールの売上
beerM = corr_beta0 + corr_beta1 * temp + stats.norm(loc=0.0, scale=0.5**2).rvs(size=1)
#print(temp[:5], beerM[:5])
#print()

# 尤度
corr_likelihood = []
for idx in range(0, len(temp)):
    beer_obs = beerM[idx]
    temp_obs = temp[idx]
    print("berr_obs: %f, temp_obs: %f" % (beer_obs, temp_obs))
    mean = corr_beta0 + corr_beta1 * temp_obs
    #mean = corr_beta0 + corr_beta1 * temp[0]
    likelihood = stats.norm.pdf(x=beer_obs, loc=mean, scale=0.5**2)
    corr_likelihood.append(likelihood)
print() 
print(corr_likelihood)

corr_likelihood = np.array(corr_likelihood) + 0.005
corr_loglikelihood = sp.log(corr_likelihood)
print()
print(corr_loglikelihood)

# 推定されたビールの売上
gbeerM = gbeta0 + gbeta1 * temp + stats.norm(loc=0.0, scale=0.5**2).rvs(size=1)
# 尤度
g_likelihood = []
for idx in range(0, len(temp)):
    gbeer_obs = gbeerM[idx]
    temp_obs = temp[idx]
    mean = beta0 + beta1 * temp_obs
    glikelibood = stats.norm.pdf(x=gbeer_obs, loc=mean, scale=0.5**2)
    g_likelihood.append(glikelibood)
print()
print(g_likelihood)
print()
print(sp.log(g_likelihood))
```
