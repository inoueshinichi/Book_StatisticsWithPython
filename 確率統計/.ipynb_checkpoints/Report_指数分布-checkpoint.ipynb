# 指数分布

$$
\begin{align}
    f(x) &= \lambda\exp(-\lambda x) \\
    E[x] &= \frac{1}{\lambda} \\
    V[x] &= \frac{1}{\lambda^{2}}
\end{align}
$$

```python
import numpy as np
import scipy as sp
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
```

確率分布

```python
x = np.linspace(0, 5, 100)
k_list = [0.1, 0.2, 0.5, 1.0, 2.0, 3.0] # k not ゼロ

for k in k_list:
    exp = stats.expon.pdf(x, scale=1.0/k)
    label = "\lambda={0:.1f}".format(k)
    plt.plot(x, exp, label=label)
plt.legend(loc='best')
plt.show()
```

サンプリング

```python
k = 2.0
e = stats.expon.rvs(scale=1.0/k, size=1000)
sns.distplot(e, bins=30, kde=False)
```
