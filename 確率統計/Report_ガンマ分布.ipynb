## ガンマ分布

$$
\begin{align}
   f(x) &= \frac{exp(\frac{-x}{\theta})}{\Gamma(k)\theta^{k}} \\
   E[x] &= k\theta \\
   V[x] &= k\theta^{2}
\end{align}
$$

```python
import numpy as np
import scipy as sp
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
```

確率密度

```python
x = np.linspace(0, 50, 100)
alpha = [1.0, 5.0, 10.0, 30.0]
beta = [1.0, 2.0, 3.0]

for b in beta:
    for a in alpha:  
        gamma = stats.gamma.pdf(x, a=a, scale=1.0/b)
        label = "alpha={0:.1f}, beta={1:.1f}".format(a,b)
        plt.plot(x, gamma, label=label)
    plt.title("Gamma")
    plt.legend(loc='best')
    plt.show()
```

サンプリング

```python
alpha = 20.0
beta  = 10.0

g = stats.gamma.rvs(a=alpha, scale=1.0/beta, size=1000)
sns.distplot(g, bins=30, kde=False)
```
