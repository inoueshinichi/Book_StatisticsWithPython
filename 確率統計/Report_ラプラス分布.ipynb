# ラプラス分布

$$
\begin{align}
    f(x) &= \frac{1}{2\sigma}\exp(-\frac{|x-\mu|}{\sigma}) \\
    E[x] &= \mu \\
    V[x] &= 2\sigma^{2}
\end{align}
$$


```python
import numpy as np
import scipy as sp
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
```

確率密度

```python
x = np.linspace(-10, 10, 100)
mean = 0
std = [1.0, 2.0, 5.0]
for sigma in std:
    laplace = stats.laplace.pdf(x, loc=mean, scale=sigma)
    label = "mean={0:.1f}, std={1:.1f}".format(mean, sigma)
    plt.plot(x, laplace, label=label)
plt.legend(loc='best')
plt.show()
```
