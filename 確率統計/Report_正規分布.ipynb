## 正規分布

$$
\begin{align}
    f(x) &= \frac{1}{\sqrt{2\pi\sigma^{2}}} \exp(-\frac{(x-\mu)^{2}}{2\sigma^{2}}) \\
    E[x] &= \mu \\
    V[x] &= \sigma^{2}
\end{align}
$$

```python
import numpy as np
import scipy as sp
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
```

確率密度

```python
x = np.linspace(-10, 10, 100)
std_list = [1.0, 2.0, 3.0, 4.0]

for std in std_list:
    norm_pdf = stats.norm.pdf(x, loc=0, scale=std)
    label = "std={0:.1f}".format(std)
    plt.plot(x, norm_pdf, label=label)
plt.xlabel("x")
plt.ylabel("p(x)")
plt.title("Normal")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

サンプリング

```python
norm = stats.norm.rvs(loc=0, scale=2, size=1000)
sns.distplot(norm, bins=30, kde=False)
```

dim = 2

```python
rng = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(rng, rng)
pos = np.dstack((xx,yy))
mean = np.array([2.5,3.3])

fig, ax = plt.subplots(2,2, figsize=(12,12))

# 左上
cov = np.array([[1.0,0.0],[0.0,1.0]])
multi_norm_pdf = stats.multivariate_normal(mean, cov).pdf(pos)
ax[0,0].contourf(xx,yy,multi_norm_pdf)
ax[0,0].set_aspect('equal')

# 右上
cov = np.array([[5.0,1.0],[1.0,2.0]])
multi_norm_pdf = stats.multivariate_normal(mean, cov).pdf(pos)
ax[0,1].contourf(xx,yy,multi_norm_pdf)
ax[0,1].set_aspect('equal')

# 左下
cov = np.array([[5.0,1.0],[2.5,2.0]])
multi_norm_pdf = stats.multivariate_normal(mean, cov).pdf(pos)
ax[1,0].contourf(xx,yy,multi_norm_pdf)
ax[1,0].set_aspect('equal')

# 右下
cov = np.array([[1.0,2.0],[0.0,5.0]])
multi_norm_pdf = stats.multivariate_normal(mean, cov).pdf(pos)
ax[1,1].contourf(xx,yy,multi_norm_pdf)
ax[1,1].set_aspect('equal')

plt.show()
```

dim=2のサンプリング

```python
mean = np.array([2.5,3.3])
cov = np.array([[5.0,1.0],[1.0,2.0]])
multi_norm = stats.multivariate_normal(mean, cov)
ys = multi_norm.rvs(size=1000)

xx,yy = np.meshgrid(rng,rng)
pos = np.dstack((xx,yy))
z = multi_norm.pdf(pos)

hist = plt.hist2d(ys[:,0], ys[:,1], 
       bins=[np.linspace(-10,10,41),np.linspace(-10,10,41)],
                  alpha=0.8)

cntr = plt.contour(xx, yy, z)
plt.axes().set_aspect('equal')
plt.title("histgram-2d")

plt.colorbar(cntr, ax=plt.axes(), shrink=0.9)
```
